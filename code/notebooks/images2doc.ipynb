```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./doc_images/six.png', cv2.IMREAD_GRAYSCALE)
```

```python
im_bw = cv2.threshold(img, 177, 255, cv2.THRESH_BINARY)[1]
```

```python
kernel = np.ones((7, 7),np.uint8)
```

```python
erosion = cv2.erode(im_bw, kernel, iterations = 1)
dilation = cv2.dilate(im_bw, kernel, iterations = 1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(im_bw, cmap='gray')
ax[0].set_title('Imagem original')
ax[1].imshow(dilation, cmap='gray')
ax[1].set_title('Dilatação')
ax[2].imshow(erosion, cmap='gray')
ax[2].set_title('Erosão')
fig.savefig(f"morph_1.pdf", bbox_inches='tight')
plt.show()
```

```python
img_op = cv2.imread('./doc_images/six_op.png', cv2.IMREAD_GRAYSCALE)
im_bw_op = cv2.threshold(img_op, 177, 255, cv2.THRESH_BINARY)[1]
opening = cv2.morphologyEx(im_bw_op, cv2.MORPH_OPEN, kernel)

fig, axes = plt.subplots(1, 2, figsize=(15, 8), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(im_bw_op, cmap='gray')
ax[0].set_title('Imagem original')
ax[1].imshow(opening, cmap='gray')
ax[1].set_title('Abertura')
fig.savefig(f"morph_2.pdf", bbox_inches='tight')
plt.show()
```

```python
img_cl = cv2.imread('./doc_images/six_cl.png', cv2.IMREAD_GRAYSCALE)
im_bw_cl = cv2.threshold(img_cl, 177, 255, cv2.THRESH_BINARY)[1]
closing = cv2.morphologyEx(im_bw_cl, cv2.MORPH_CLOSE, kernel)

fig, axes = plt.subplots(1, 2, figsize=(15, 8), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(im_bw_cl, cmap='gray')
ax[0].set_title('Imagem original')
ax[1].imshow(closing, cmap='gray')
ax[1].set_title('Fechamento')
fig.savefig(f"morph_3.pdf", bbox_inches='tight')
plt.show()

```
