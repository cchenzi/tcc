```python
def print_image(images, subtitles, title):
    fig, axes = plt.subplots(1, len(images), figsize=(20, 15), sharex=True, sharey=True)
    ax = axes.ravel()
#     fig.suptitle("test")
    for idx, a in enumerate(ax):
        a.get_xaxis().set_visible(False)
        a.get_yaxis().set_visible(False)
        ax[idx].imshow(images[idx])
#     ax[0].set_title(subtitles[0])
#     ax[1].imshow(images[1])
# #     ax[1].set_title(subtitles[1])
#     ax[2].imshow(images[2])
#     ax[2].set_title(subtitles[2])
    print(title)
    fig.savefig(f"{title}.pdf", bbox_inches='tight')
    plt.show()
```

```python
import pytesseract
```

```python
def ocr_to_boxes(filename):
    img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    print(pytesseract.image_to_string(img))
    d = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
    vis = cv2.imread(filename, 1)
    vis = cv2.cvtColor(vis, cv2.COLOR_BGR2RGB)
    n_boxes = len(d['text'])
    for i in range(n_boxes):
        if int(d['conf'][i]) > 60:
            (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
            vis = cv2.rectangle(vis, (x, y), (x + w, y + h), (255, 255, 0), 5)
    return vis
#     plt.figure(figsize=(220,14))
#     plt.imshow(vis)
```

```python
a = ocr_to_boxes('../images/agua/IMG_20210312_090250907.jpg')
```

```python
fig = plt.figure(figsize=(220,14))
plt.imshow(a)
# fig.savefig(f"test.pdf", bbox_inches='tight')
```
